```python
import pandas as pd
import numpy as np
import json
```

```python
df = pd.read_excel("Problem_C_Data_Wordle.xlsx",usecols=range(1,13))
df
```

```python
oneletter_list = ["e","t","a","o","i"]
twoletter_list = ["th","he","in","er","an"]
word_list = df["Word"]
```

```python
for oneletter in oneletter_list:
    list = []
    for i in range(len(df)):
        list.append(df["Word"][i].count(oneletter))
    df[oneletter] = pd.Series(list)
for twoletter in twoletter_list:
    list = []
    for i in range(len(df)):
        list.append(df["Word"][i].count(twoletter))
    df[twoletter] = pd.Series(list)
df
```

```python
with open("英文单词词频表.json","r") as f:
    content = f.read()
    frequency_table = json.loads(content)
```

```python
list = []
for i in range(len(df)):
    list.append(frequency_table[df["Word"][i]])
df["Frequency"] = pd.Series(list)
df
```

```python
df.to_csv("Wordle表格含单词频率.csv")
```
